```python
import pandas as pd

import ast
def tokens_to_list(text):
    text = text[1:-1]
    res = ast.literal_eval(text)
    return list(dict(res).keys())


clusters = pd.read_csv("../../data/bertopic.csv")
clusters = clusters[1:]
clusters.head()
```

```python
clusters['tokens'] = clusters.most_freq_tokens.apply(lambda x: tokens_to_list(x))
clusters
```

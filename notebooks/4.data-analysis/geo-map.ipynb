```python
import folium
from folium.plugins import HeatMap, HeatMapWithTime
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
data = pd.read_csv("../../data/loslagos-coordinates.csv")
```

```python
from random import randrange
from datetime import timedelta
from datetime import datetime

# create multiple rows 
data = pd.concat([data]*20, ignore_index=True)

# create random values
data['Values'] = np.random.rand(len(data))
```

```python
# create random dates
def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)


d1 = datetime.strptime('1/1/2021', '%d/%m/%Y')
d2 = datetime.strptime('30/1/2021', '%d/%m/%Y')
dates = []
for i in range(len(data.City)):
    dates.append(str(random_date(d1, d2)).split(" ")[0])
data['Date'] = dates
```

```python
data=data.sort_values(by=['Date'])
time_index = list(data['Date'].sort_values().astype('str').unique())
```

```python
what = []
for _, d in data.groupby('Date'):
    what.append([[row['Latitude'], row['Longitude'], row['Values']] for _, row in d.iterrows()])
```

```python
hmt = folium.Map(location=[-41.4693, -72.94237],
               tiles='cartodbpositron',#'cartodbpositron', stamentoner
               zoom_start=7,
               control_scale=True)
HeatMapWithTime(what,
                index=time_index,
                auto_play=True,
                radius=60,
                position='topright',
                #use_local_extrema=True,
                #min_opacity=0.4,
                #blur = 0.2
               ).add_to(hmt)
folium.LayerControl().add_to(hmt)

hmt
```
